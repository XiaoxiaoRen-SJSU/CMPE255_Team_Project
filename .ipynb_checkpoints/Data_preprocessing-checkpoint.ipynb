# Label dataset

## ipo and acquired, lable = 1, others label = 0

```python
companies.groupby('status').count()
```

```python
companies_with_numeric_label = companies['label'] = 0
companies.loc[companies['status'] == 'ipo', 'label'] = 1
companies.loc[companies['status'] == 'acquired', 'label'] = 1
companies.head()
```

```python
investments_num = investments[['company_permalink', 'investor_permalink']].groupby(['company_permalink']).agg(['count'])
investments_num.columns = ['Num_of_investor']
investments_num.head()
```

```python
comp_plus_InvestorNum = pd.merge(how='left',left=companies, right=investments_num, left_on='permalink', right_on='company_permalink')
comp_plus_InvestorNum.head()
```

```python
comp_plus_InvestorNum['funding_total_usd'] = pd.to_numeric(comp_plus_InvestorNum['funding_total_usd'], errors='coerce')
```

```python
comp_plus_InvestorNum.head()
```

```python
t1 = pd.to_datetime(comp_plus_InvestorNum.first_funding_at, errors='coerce')
t1 = pd.to_timedelta(t1).dt.days
t2 = pd.to_datetime(comp_plus_InvestorNum.last_funding_at, errors='coerce')
t2 = pd.to_timedelta(t2).dt.days
```

```python
comp_plus_InvestorNum['funding_duration'] = t2 - t1
comp_plus_InvestorNum.head()
```

```python
comp_plus_InvestorNum['first_funding_at_UTC'] = t1
comp_plus_InvestorNum['last_funding_at_UTC'] = t2
```

```python
companies_allFeature = comp_plus_InvestorNum
```
